```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline 
df = pd.read_csv('HR_comma_sep.csv')
df.head()
```

```python
dummiesDep = pd.get_dummies(df["Department"])
df_dum = pd.concat([df,dummiesDep],axis="columns")
df_dum = df_dum.drop(["Department"],axis="columns")
df_dum.head()
```

```python
dummiesSal = pd.get_dummies(df_dum["salary"])
df_sal = pd.concat([df_dum,dummiesSal],axis="columns")
df_sal = df_sal.drop(["salary"],axis="columns")
df_sal.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(df_sal[['satisfaction_level', 'last_evaluation','number_project','average_montly_hours','time_spend_company','Work_accident','promotion_last_5years','IT','RandD','accounting','hr','management','marketing','product_mng','sales','support','technical','high','low','medium']],df.left,test_size=0.1)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```
